# Summary Statistics
Compute various summary statistics for the DataFrame.

```python
# Summary Statistics

# Select only numeric columns
numeric_df = netflix_df.select_dtypes(include='number')

# Compute mean of columns
mean_values = numeric_df.mean()
print("Mean values:\n", mean_values)

# Compute median of columns
median_values = numeric_df.median()
print("Median values:\n", median_values)

# Compute mode of columns
mode_values = numeric_df.mode()
print("Mode values:\n", mode_values)

# Compute minimum of columns
min_values = numeric_df.min()
print("Minimum values:\n", min_values)

# Compute maximum of columns
max_values = numeric_df.max()
print("Maximum values:\n", max_values)

# Compute count of columns
count_values = numeric_df.count()
print("Count values:\n", count_values)

# Compute variance of columns
variance_values = numeric_df.var()
print("Variance values:\n", variance_values)

# Compute standard deviation of columns
std_dev_values = numeric_df.std()
print("Standard deviation values:\n", std_dev_values)

# Compute sum of columns
sum_values = numeric_df.sum()
print("Sum values:\n", sum_values)

# Apply multiple functions to columns using agg()
agg_values = numeric_df.agg(['min', 'max', 'mean', 'median'])
print("Aggregated values:\n", agg_values)

# Compute cumulative sum of columns
cumsum_values = numeric_df.cumsum()
print("Cumulative sum values:\n", cumsum_values)

# Compute cumulative maximum of columns
cummax_values = numeric_df.cummax()
print("Cumulative maximum values:\n", cummax_values)
```

# Filter Movies Released in the 1990s
Filter the DataFrame to include only movies released in the 1990s.

```python
# Filter Movies Released in the 1990s
# Filter the DataFrame to include only movies released in the 1990s.

# Subset the DataFrame for type "Movie"
netflix_subset = netflix_df[netflix_df["type"] == "Movie"]

# Filter the DataFrame to keep only movies released in the 1990s
movies_1990s = netflix_subset[(netflix_subset["release_year"] >= 1990) & (netflix_subset["release_year"] < 2000)]
movies_1990s
```

# Filter Action Movies in the 1990s
Filter the DataFrame to include only action movies released in the 1990s.

```python
# Filter Action Movies in the 1990s
# Filter the DataFrame to include only action movies released in the 1990s.

# Filter the DataFrame to keep only movies released in the 1990s
movies_1990s = netflix_subset[(netflix_subset["release_year"] >= 1990) & (netflix_subset["release_year"] < 2000)]

# Filter the data again to keep only the Action movies
action_movies_1990s = movies_1990s[movies_1990s["genre"] == "Action"]
action_movies_1990s
```

# Count Short Action Movies in the 1990s
Count the number of short action movies (duration < 90 minutes) released in the 1990s.

```python
# Count Short Action Movies in the 1990s
# Count the number of short action movies (duration < 90 minutes) released in the 1990s.

# Start the counter
short_movie_count = 0

# Iterate over the labels and rows of the DataFrame and check if the duration is less than 90, if it is, add 1 to the counter, if it isn't, the counter should remain the same
for label, row in action_movies_1990s.iterrows():
    if row["duration"] < 90:
        short_movie_count += 1

short_movie_count

# A quicker way of counting values in a column is to use .sum() on the desired column
short_movie_count_quick = (action_movies_1990s["duration"] < 90).sum()
short_movie_count_quick
```
